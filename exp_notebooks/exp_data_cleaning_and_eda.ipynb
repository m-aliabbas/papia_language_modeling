## Data Cleaning

```python
# !pip install pandas
```

```python
import pandas as pd
```

```python
import re

def clean_text(text):
    # Function to clean text
    # Remove URLs
    text = re.sub(r'http\S+|www.\S+', '', text)
    # Remove all the special characters
    text = re.sub(r'\W', ' ', text)
    # Remove all single characters
    text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text)
    # Remove single characters from the start
    text = re.sub(r'\^[a-zA-Z]\s+', ' ', text) 
    # Substituting multiple spaces with single space
    text = re.sub(r'\s+', ' ', text, flags=re.I)
    # Removing prefixed 'b'
    text = re.sub(r'^b\s+', '', text)
    # Convert to Lowercase
    text = text.lower()
    return text
```

```python
# Function to remove lines that only contain one word or character
def remove_single_word_lines(text):
    # Split text into lines
    lines = text.split('\n')
    # Only keep lines that contain more than one word
    new_lines = [line for line in lines if len(line.split()) > 1]
    # Join lines back into text
    text = '\n'.join(new_lines)
    return text
```

```python
import re

def clean_text_file(input_file_path, output_file_path=''):
    # Read the text file
    with open(input_file_path, 'r') as f:
        data = f.read()
    # Clean the data
    cleaned_data = clean_text(data)
    # Remove single word lines
    cleaned_data = remove_single_word_lines(cleaned_data)
#     # Write the cleaned data back into a text file
#     with open(output_file_path, 'w') as f:
#         f.write(cleaned_data)
    return cleaned_data
```

```python
data_dict = {file_name:clean_text_file(file_name) for file_name in file_names}
```

```python
file_names = data_dict.keys()
clean_data = data_dict.values()
```

```python
clean_data_dict = {'file_names':file_names,'data':clean_data}
```

```python
df = pd.DataFrame(clean_data_dict)
```
